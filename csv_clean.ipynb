```python
import pandas as pd
import numpy as np
```

## Clean caracteristiques csv

```python
for i in range(2006,2022):
    if i==2009:
        delimiter = "\t"
    elif i >= 2019:
        delimiter = ";"
    else:
        delimiter = ","

    df = pd.read_csv(f'./accidents/caracteristiques_{i}.csv', delimiter=delimiter, encoding='latin-1')
    
    # drop adr column
    df.drop('adr', axis=1, inplace=True)

    if i<2019:

        # remove the last zero of dep column
        df['dep'] = df['dep'].apply(lambda x: str(x)[:-1].zfill(2) if str(x) != '2A' and str(x) != '2B' and str(x)[-1]=="0" else str(x))

        
        # if lenght(com) <3 add 0 at the begining
        df['com'] = df['com'].apply(lambda x: str(x).zfill(3) if len(str(x))<3 else str(x))

        # cooncatenate dep and com
        df['com'] = df['dep'] + df['com']

        # if len(com) >5 remove the 2nd car
        df['com'] = df['com'].apply(lambda x: str(x)[:2] + str(x)[3:] if len(str(x))>5 else str(x))

    if i >= 2019:
        # transform an from 4 digit to 2 digit
        df['an'] = df['an'].apply(lambda x: str(x)[-2:])
    
    # compute gps
    df['gps'] = df['dep'].apply(
        lambda x: "M" if "01" <= str(x) <= "95" 
            else "A" if str(x)=="971" or str(x)=="972" 
            else "G" if str(x)=="973" 
            else "R" if str(x)=="974" 
            else "S" if str(x)=="975" 
            else "Y" if str(x)=="976" 
            else "T" if str(x)=="984" 
            else "P" if str(x)=="987" 
            else "W" if str(x)=="986" 
            else "C" if str(x)=="988" 
            else np.nan)

    # reorganize columns
    df = df[['Num_Acc', 'an', 'mois', 'jour', 'hrmn', 'lum', 'agg', 'int', 'atm', 'col', 'com', 'gps', 'lat', 'long', 'dep']]


    df.to_csv(f'./accidents2/caracteristiques_{i}.csv', index=False)
```

```python
# open all csv in accidents2 folder and concat them
df = pd.concat([pd.read_csv(f'./accidents2/caracteristiques_{i}.csv') for i in range(2006,2022)], ignore_index=True)
df
```
